## 2 - Etapa de validação

```python
### Parametro dayfirst informa que no arquivo original o primeiro parametro é o dia e não o mes.
df = pd.read_csv("ocorrencia_2010_2020.csv", delimiter=';', parse_dates=['ocorrencia_dia'], dayfirst=True)
df.head()
```

```python
df.dtypes
```

```python
df.ocorrencia_dia.dt.month
```

```python
df.head(10)
```

```python
## pa.Check.str_matches, utiliza uma expressão regular para validar o formato de hora (utilização é opcional)
## Opção required para campos que não são obrigatorios vir na base de dados.

schema = pa.DataFrameSchema(
    columns = {
        "codigo": pa.Column(pa.Int, required=False),
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2)),
        "ocorrencia_aerodromo": pa.Column(pa.String),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)
    }
)
```

```python
### Caso o comando abaixo de erro, o schema de validação acima, apontou erros/diferenças.
schema.validate(df)
```

## 4 - Etapa de Transformação

```python
valores_ausentes = ['****','###!','####','**','*****','NULL']
df = pd.read_csv("ocorrencia_2010_2020.csv", delimiter=';', parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head()
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo": pa.Column(pa.Int, required=False),
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable = True),
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable = True),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)
    }
)
```

```python
## OCORRENCIAS QUE COMEÇAM COM C
filtro = df.ocorrencia_cidade.str[0] == 'C'
df.loc[filtro, ['ocorrencia_cidade', 'ocorrencia_uf']]
```

```python
## OCORRENCIAS QUE TERMINAM COM A
filtro = df.ocorrencia_cidade.str[-1] == 'A'
df.loc[filtro, ['ocorrencia_cidade', 'ocorrencia_uf']]
```

```python
## OCORRENCIAS QUE TERMINAM COM NA
filtro = df.ocorrencia_cidade.str[-2:] == 'NA'
df.loc[filtro, ['ocorrencia_cidade', 'ocorrencia_uf']]
```

```python
## OCORRENCIAS QUE CONTÉM MA
filtro = df.ocorrencia_cidade.str.contains('MA')
df.loc[filtro, ['ocorrencia_cidade', 'ocorrencia_uf']]
```

```python
## OCORRENCIAS QUE CONTÉM MA ou AL
filtro = df.ocorrencia_cidade.str.contains('MA|AL')
df.loc[filtro, ['ocorrencia_cidade', 'ocorrencia_uf']]
```

```python
## OCORRENCIAS DO ANO DE 2015
filtro = df.ocorrencia_dia.dt.year == 2015
df.loc[filtro, ['ocorrencia_cidade', 'ocorrencia_uf']]
```

```python
## OCORRENCIAS DO ANO DE 2015 e entre os dias 3 e 8 de cada mes
filtro = (df.ocorrencia_dia.dt.year == 2015) & ((df.ocorrencia_dia.dt.day >=3) & (df.ocorrencia_dia.dt.day <= 8))
df.loc[filtro, ['ocorrencia_cidade', 'ocorrencia_uf', 'ocorrencia_dia']]
```

```python
## criar coluna de dia e hora
df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
```
